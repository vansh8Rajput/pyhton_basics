Q14. Describe a scenario where a while loop is more suitable than a for loop.
Ans. Scenario: User Login System
Imagine you're building a program that keeps asking a user to enter the correct password until they get it right.
In this case, you don’t know how many attempts it will take, so a while loop is perfect:

correct_password = "jammy123"
user_input = input("Enter your password: ")
while user_input != correct_password:
    print("Wrong password. Try again!")
    user_input = input("Enter your password: ")
print("Access granted!")

Why while is better here?
Because the number of iterations is not fixed.
You loop until a condition is met (correct password).
A for loop is better when you know the number of times you want to loop, like looping through a list or range.

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number to calculate its factorial: "))
factorial = 1

# Check if the entered number is negative or zero
if num < 0:
    print("Factorial does not exist for negative numbers")
else:
    while num > 1:
        factorial *= num  # Multiply factorial with the current number
        num -= 1  # Decrease the number by 1

    print("The factorial is:", factorial)
```
